# 과제제출

## 폴더 구성

* 수업시간에 하는대로 폴더를 구성한다.
* 폴더명은 영문으로 한다.
* 배정받은 자신의 폴더가 작업디렉토리working directory가 된다.
* 패키지는 분반, 주차, 문제 번호
* 폴더와 패키지 구성 예 package com.j2.w1.p1
    ```
    classes/
    src/
        com/
            j2/
                w1/
                    p1/
    ```

## 파일 명명

* 파일명은 영어로 작성한다.
* 파일명은 
    * 과목명 영어이니셜 1자
    * 분반 숫자 1자 
    * 학번 
    * 주차 2자
* 메인함수가 있는 경우, 맨 뒤에 main을 붙인다.
* class는 올리지 않는다.
    * 텍스트 파일명 예제
    ```
        j2_201399999_w2.txt
        j2_201399999_w2main.java
    ```

            // comment
            public class j2_201399999_w2main {

                    p1main();
                    p2main();

                    // comment

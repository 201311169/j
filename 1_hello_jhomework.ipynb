# 과제제출

## 폴더 구성

* 폴더명은 영문으로 한다.
* 작업디렉토리 (wd, working directory)
    * wd는 .git가 있는 디렉토리이다.
    * github.com에서 자신의 계정에 생성한 j1_201399999 리포를 git clone하면 된다.
    * wd 명명 - 자신이 누구인지 식별할 수 있게, 분반과 학번을 사용한다.
        * 과목명 영어이니셜 1자
        * 분반 숫자 1자 
        * 학번
        * 예: j1_201399999
    * wd의 하위폴더는 아래와 같이 구성한다 (수업시간에 했던 그대로. 아래 참조)
    * 패키지명에서 분반, 주차, 문제 번호를 알 수 있게 한다.
* 패키지
    * 패키지는 영문으로 한다.
    * wd와 패키지 구성 예 -> package com.j1.w1.p1
    ```
    j1_201399999
        doc/
        classes/
        src/
            com/
                j1/
                    w1/
                        p1/
    ```
    
* 파일
    * 파일명은 영어로 작성한다.
    * 파일 명명 - 주차와 메인함수가 있는 파일을 알 수 있게 한다.
        * 메인함수가 있는 경우, 맨 뒤에 Main을 붙인다.
        * 텍스트 파일은 md을 사용한다.
        ```
        w2.md
        w2Main.java
        ```

## git로 과제제출

* git server
  * github.com (bitbucket.org 또는 Google Code를 사용할 수 있다.)
  * 리포를 만든다 j1\_201399999 repository
  * 자신의 계정을 공유한다 (수업시간에 방법을 공지 한다.)
  * 기한에 commit해야 한다.

* .gitignore
  * class는 올리지 않는다.
  * jar는 올리지 않는다.
  * java~는 올리지 않는다.
  * 그 외 바이너리 파일은 올리지 않는다.

## 채점기준

-   문제에 적합한 정상실행 100%
-   오작동 미제출 0점 처리.
-   제출시점 미준수: 24시간 내 제출 50% 감점. 24시간 이후 0점 처리.
-   모든 Lab은 반드시 제출. 미제출 비율은 성적에 반영 (예: 20% 미제출
    최대 성적은 B로 함)

## 프로그램 형식

* 설명은 // comment로 작성 (주의: 영어로 작성)
* 예
    ```
    package com.j1.w1;

    import com.j1.wk1.p1.*;
    import com.j1.wk1.p2.*;
    import java.awt.Point;
    import java.util.ArrayList;
    import java.util.Arrays;

    // comment
    public class w2Main {
        public static void main(String[] args) {
            p1main();
            p2main();
        }
        // comment
        private static void p1main() {
            // comment
            System.out.println("--p1--");
        }
        private static void p2main() {
            System.out.println("--Rectangle Square");
            System.out.println("--association");
            System.out.println("--sorter");
        }
    }             
    ```
    
* 과제 테스트: ant 또는 java
    * java
    ```
    $ javac -d classes/ -sourcepath src/ src/com/j1/w1/w2Main.java
    $ java -cp classes/ com.j1.w1.w2Main        
    ```
